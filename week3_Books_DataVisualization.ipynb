## Books Dataset Visualizations

```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Books.csv')
df.head()
```

```python
df.describe()
df.info()
```

```python
print("Dataset Shape:", df.shape)
```

```python
print(df.columns)
```

```python
print('\nMissing Values:')
print(df.isnull().sum())
```

#### There are no missing values present in the dataset

```python
numeric_df = df.select_dtypes(include=[np.number])
print('\nNumeric Columns:', list(numeric_df.columns))
```

```python
top_titles = df['title'].value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(y=top_titles.index, x=top_titles.values)
plt.title('Top 10 Most Frequent Book Titles')
plt.xlabel('Count')
plt.ylabel('Book Title')
plt.show()
```

```python
# Convert to numeric, remove invalid years
df['published_date'] = pd.to_numeric(df['published_date'], errors='coerce')
df_clean = df[df['published_date'].between(1900, 2025)]

plt.figure(figsize=(12,5))
sns.histplot(df_clean['published_date'], bins=50, kde=False)
plt.title('Number of Books Published Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Books')
plt.show()
```

```python
top_publishers = df['publisher'].value_counts().head(5)
top_publishers.plot.pie(autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Publishers by Book Count')
plt.ylabel('')
plt.show()
```

```python
top_publishers = df['publisher'].value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(y=top_publishers.index, x=top_publishers.values)
plt.title('Top 10 Publishers by Book Count')
plt.xlabel('Number of Books')
plt.ylabel('Publisher')
plt.show()
```

```python
sns.histplot(df['ratings_count'], bins=50, kde=True)
plt.title('Distribution of Ratings Count')
plt.xlabel('Ratings Count')
plt.ylabel('Number of Books')
plt.show()
```

```python
sns.boxplot(x=df['ratings_count'])
plt.title('Boxplot of Ratings Count')
plt.xlabel('Ratings Count')
plt.show()
```

### Bivariate Analysis Visualizations  :

```python
df.head()
```

```python
top_authors = df['author'].value_counts().head(10).index
author_avg = df[df['author'].isin(top_authors)].groupby('author')['ratings_count'].mean().sort_values(ascending=False)

sns.barplot(x=author_avg.values, y=author_avg.index)
plt.title('Top Authors by Average Ratings Count')
plt.xlabel('Average Ratings Count')
plt.ylabel('Author')
plt.show()
```

```python
df['published_date'] = pd.to_numeric(df['published_date'], errors='coerce')
df_clean = df[df['published_date'].between(1900, 2025)]

sns.scatterplot(x='published_date', y='ratings_count', data=df_clean)
plt.title('Ratings Count Over Publication Years')
plt.xlabel('Year of Publication')
plt.ylabel('Ratings Count')
plt.show()
```

```python
top_pubs = df['publisher'].value_counts().head(5).index
df_top_pubs = df[df['publisher'].isin(top_pubs)]

sns.boxplot(x='publisher', y='ratings_count', data=df_top_pubs)
plt.title('Ratings Count by Publisher')
plt.xlabel('Publisher')
plt.ylabel('Ratings Count')
plt.xticks(rotation=45)
plt.show()
```

```python
df['pages'] = pd.to_numeric(df['pages'], errors='coerce')

df_cleaned = df.dropna(subset=['pages'])

plt.figure(figsize=(10, 6))
sns.histplot(df_cleaned['pages'], bins=50, kde=True, color='skyblue')
plt.title('Distribution of Number of Pages in Books')
plt.xlabel('Number of Pages')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.tight_layout()
```
